# Imports

```python
import matplotlib.pyplot as plt
import os
import numpy as np
```

# Function

```python
def display_images_in_rows(image_paths, images_per_row=4):
  """Displays a list of images in rows with a specified number of images per row.

  Args:
    image_paths: A list of paths to the images to display.
    images_per_row: The number of images to display in each row.
  """
  num_images = len(image_paths)
  num_rows = (num_images + images_per_row - 1) // images_per_row

  for row in range(num_rows):
    start_index = row * images_per_row
    end_index = min((row + 1) * images_per_row, num_images)
    row_images = image_paths[start_index:end_index]

    plt.figure(figsize=(15, 5))  # Adjust figure size as needed
    for i, image in enumerate(row_images):
      plt.subplot(1, images_per_row, i + 1)
      plt.imshow(image)
      plt.axis('off')

    plt.show()

def adjust_contrast(image, contrast_factor=0.5):
  """Adjusts the contrast of an image.

  Args:
    image: The input image as a NumPy array.
    contrast_factor: The factor to adjust the contrast by. A value greater than
      1 increases contrast, while a value less than 1 decreases contrast.

  Returns:
    The image with adjusted contrast.
  """
  mean = np.mean(image)
  adjusted_image = (image - mean) * contrast_factor + mean
  adjusted_image = np.clip(adjusted_image, 0, 255).astype(np.uint8)
  return adjusted_image
```

```python
dir_path = "./Dataset/raw/yes"
image_paths =  os.listdir(dir_path)
image_paths
```

```python
image_paths = [os.path.join(dir_path, path) for path in image_paths]
image_paths
```

```python
images = []

for path in image_paths:
    images.append(plt.imread(path))
```

```python
contrast_image = []

for image in images:
    contrast_image.append(adjust_contrast(image, 1.25))
```

```python
display_images_in_rows(contrast_image[:20], 10)
```

```python
for i in range(len(contrast_image)):
    plt.imsave(os.path.join("./Dataset/processed/yes", f'y{i}.jpg'), contrast_image[i])
```
